```python
def sliced_data (data,start_key,end_key):
# List all keys in the .mat file
    keys = list(data.keys())

# Define the start and end key
    #start_key = 'Hsig_97010100_0'
    #end_key = 'Hsig_98123100_0'

# Filter the keys that fall within the range from 'Hsig_97010100_0' to 'Hsig_98123100_0'
    selected_keys = [key for key in keys if start_key <= key <= end_key]

# Slice the data for the selected keys
    sliced_mat_data = {key: data[key] for key in selected_keys}
    return sliced_mat_data

# Example: Accessing data from a specific key
#for key, data in sliced_data.items():
   # print(f"Key: {key}, Data: {data}")
```

```python
# Calculating two yearly ( 2021 -2022) & 2097 -2098 average over domain  

def calculate_average_hsig(mat_data):
    # Step 1: Extract all Hsig arrays from mat_data (ignoring keys that start with '__')
    hsig_data = []
    for key, value in mat_data.items():
        if key.startswith('Hsig_'):
            hsig_data.append(value)
    
    # Step 2: Stack the arrays along a new axis (time axis)
    hsig_stack = np.stack(hsig_data, axis=0)  # Shape will become (time, lat, lon)

    # Step 3: Calculate the mean, ignoring NaN values
    avg_hsig = np.nanmean(hsig_stack, axis=0)  # Average over the time dimension

    return avg_hsig

```

```python
# Assuming 'mat_data' is your dictionary with Hsig arrays
winter_avg_P = calculate_average_hsig(WR_P)
winter_avg_F = calculate_average_hsig(WR_F)
spring_avg_P=calculate_average_hsig(SG_P)
spring_avg_F=calculate_average_hsig(SG_F)
summer_avg_P=calculate_average_hsig(SR_P)
summer_avg_F=calculate_average_hsig(SR_F)
autumn_avg_P=calculate_average_hsig(AM_P)
autumn_avg_F=calculate_average_hsig(AM_F)
```
